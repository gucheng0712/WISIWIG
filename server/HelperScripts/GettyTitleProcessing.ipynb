```python
import pandas as pd
from pandas import ExcelFile
df = pd.read_excel('getty_dataset/TECH 2709 Getty Dataset.xlsx')
```

```python
print(df.head())
```

```python
df.drop(df.index[0:4],axis=0)
```

```python
df_descriptions = df['Title']
df_descriptions.count()
```

```python
res = []
for d in df_descriptions.values:
    if (not pd.isnull(d)):
        d = d.replace('\n','')
        d = d.strip()
        res.append(d)
res.pop(0)
print(res)
# descriptions = []
# for i in range(df_descriptions.count()):
#     d = df_descriptions[df_descriptions.index[i]]
#     if (not pd.isnull(d)):
#         d = d.replace('\n','')
#         d = d.strip()
#         descriptions.append(d)
# print(descriptions)
```

```python
with open('getty_dataset/getty-titles.utf8.txt','w') as filehandle:
    for d in res:
        filehandle.write('%s\r'%d)
print("wrote file..")
```

```python
text_file = open("getty_dataset/getty-titles.utf8.txt", "r")
titles = text_file.read().split('\n')
titles
```

```python
from textgenrnn import textgenrnn
textgen = textgenrnn()
textgen.train_on_texts(titles, num_epochs=10)
textgen.save('getty-titles-model-e10.hdf5')
```

```python
# comment out below to read a saved model:
textgen = textgenrnn('getty-titles-model-e10.hdf5')
gem_titles = []

for i in range(30):
    gen_titles+=textgen.generate(10, temperature=0.5, return_as_list=True)

for t in gen_titles:
    if t in titles:
        gen_titles.remove(t)
gen_titles = list(set(gen_titles))

print(gen_titles)
```
